```python
# lets calculate the cosine similarity between all writers
author_list = writer_features['author_fn'].tolist()
# preallocate the matrix
all_cos_sim = np.ones([len(author_list), len(author_list)])
for curr_name in author_list:
    # get suggestions from each author
    curr_matches_df = s.give_author_suggestion_from_author(writer_features, curr_name, nshow=len(author_list))
    curr_similarities = curr_matches_df['Similarity (0-10)'].as_matrix()
    curr_names_sorted = curr_matches_df['Authors'].tolist()
    # resort to stick back in matrix
    resorted_list = []
    for i in author_list:
        if i == curr_name:
            resorted_list.append(100)
        elif curr_similarities[curr_names_sorted.index(i)] < 0:
            resorted_list.append(0)           
        else:
            resorted_list.append(curr_similarities[curr_names_sorted.index(i)]**2)
    # putting in matrix
    all_cos_sim[author_list.index(curr_name),:] = resorted_list
```

```python
plt.imshow(all_cos_sim)
plt.colorbar()
plt.show()
# np.log10(10)
```

```python
# now lets make a graph
G = nx.DiGraph()
G.add_nodes_from(author_list)
for i in range(0,len(author_list)):
    for j in range(i,len(author_list)): 
        if all_cos_sim[i,j] > 0:
            G.add_edge(author_list[i], author_list[j], weight=np.max([all_cos_sim[i,j], 0]))
        else:
            G.add_edge(author_list[i], author_list[j], weight=1)
```

```python
# for visualization
# pos = nx.kamada_kawai_layout(G)
# pos=nx.spring_layout(G, iterations=500)
pos=nx.circular_layout(G)
cm = plt.get_cmap('inferno_r', 7) # 7 bc of the 5 positions and no black
unique_pos = ['PG', 'SG', 'SF', 'PF', 'C']
curr_positions = nx.get_node_attributes(G, 'pos')
nodecolors = [cm.colors[j] for i in curr_positions for j in range(0,len(unique_pos))
              if curr_positions[i] == unique_pos[j]]
edge_labels=dict([((u, v, ), round(d['weight']))
                 for u, v, d in G.edges(data=True)])
# for the node size
node_size_use = nx.get_node_attributes(G, 'ppg')
node_array = [node_size_use[i]*25 for i in list(node_size_use)]
labels_use = {}
for i in G.nodes():
    labels_use[i] = i
# print(edge_labels)
edgewidth = [d['weight']/40 for (u, v, d) in G.edges(data=True)]
plt.figure(figsize=(10, 4))
# # we can now added edge thickness and edge color
nx.draw_networkx_nodes(G, pos=pos, alpha=0.8)
nx.draw_networkx_edges(G, pos, width=edgewidth, alpha=0.15, arrowstyle='simple',
                       arrowsize=5, arrows=False, edge_cmap=plt.cm.Greys)
nx.draw_networkx_labels(G, pos, labels_use, font_size=8)
plt.axis('off')
plt.tight_layout()
```

```python
# load necessary models and keyed vectors and initialize mongo db
# load word2vec model
kv = gensim.models.KeyedVectors.load(
    "/Users/rohanramesh/Documents/Insight/data_bball_writers/word2vec_model_kv.kv", mmap='r')
# load word2vec df for comparisons
w2v_df = pd.read_pickle('/Users/rohanramesh/Documents/Insight/data_bball_writers/word2vec_trained.pickle')

client = MongoClient()
mydb = client["testinsightdb"]
```

```python
# detect if no english and return a flag
def detect_nonenglish(txtstr):
    b = sent_tokenize(txtstr)
    b = b[0:np.max([10,len(b)])] # take longest of first 10 sentences
    # n words per sentence
    nwords = [len(word_tokenize(i)) for i in b]
    if not nwords:
        output = True
    else:
        idx = nwords.index(np.max(nwords))
        c = detect(b[idx])
        if c != 'en':
            output = True
        else:
            output = False
    return output
```

```python
# lets take same approach but with top 250 articles for each person
author_list = writer_features['author_fn'].tolist()
article_order = w2v_df.keys().tolist()
# content_similarities = []
# style_similarities = []
# urls = []
# authors_per_article = []
for ii in range(75,len(author_list)):
    curr_name = author_list[ii]
# for curr_name in author_list:
    # convert to website author name
    print(curr_name)
    website_name = s.match_author_names(writer_df, curr_name)
    currcol = mydb[website_name]
    y = currcol.find({"name": website_name})
    idx = 1
    for curr_art in y:
        # check if english
        a = detect_nonenglish(curr_art['article'])
        if a == True:
            continue
        if idx > 250:
            print('breaking')
            break
        idx += 1
        # for content similarity
        lem_art = s.lemstr(gensim.utils.simple_preprocess(curr_art['article']))
        curr_matches_df_content = s.recommend_article_content(kv, w2v_df, lem_text=lem_art, article_url=None)
        curr_similarities = curr_matches_df_content['Similarity (0-10)'].as_matrix()
        curr_url_sorted = curr_matches_df_content['Suggested articles'].tolist()
        # resort to stick back in matrix
        resorted_list_content = []
        for i in article_order:
            try:
                resorted_list_content.append(curr_similarities[curr_url_sorted.index(i)])
            except:
                resorted_list_content.append(-1)
        content_similarities.append(resorted_list_content)
        # for style similarity
        curr_matches_style = s.give_suggestion_featurespace_single_article(writer_features, 
                                                          txtstr=curr_art['article'], nshow=len(author_list))
        curr_similarities = curr_matches_style['Similarity (0-10)'].as_matrix()
        curr_names_sorted = curr_matches_style['Authors'].tolist()
        # resort to stick back in matrix
        resorted_list_style = []
        for i in author_list:
            if i == curr_name:
                resorted_list_style.append(10)         
            else:
                resorted_list_style.append(curr_similarities[curr_names_sorted.index(i)])
        style_similarities.append(resorted_list_style)
        # save url
        urls.append(curr_art['article_url'])
        authors_per_article.append(curr_name)
```

```python
# lets just save entire list
all_article_cos_sim__style_content_url_authors = [content_similarities,
                                                 style_similarities,
                                                 urls,
                                                 authors_per_article]
with open('/Users/rohanramesh/Documents/Insight/data_bball_writers/all_articles_cossim_styleContentUrlAuthors.pickle', 'wb') as handle:
    pickle.dump(all_article_cos_sim__style_content_url_authors, handle, protocol=pickle.HIGHEST_PROTOCOL)
```
