```python
# lets calculate the cosine similarity between all writers
author_list = writer_features['author_fn'].tolist()
# preallocate the matrix
all_cos_sim = np.ones([len(author_list), len(author_list)])
for curr_name in author_list:
    # get suggestions from each author
    curr_matches_df = s.give_author_suggestion_from_author(writer_features, curr_name, nshow=len(author_list))
    curr_similarities = curr_matches_df['Similarity (0-10)'].as_matrix()
    curr_names_sorted = curr_matches_df['Authors'].tolist()
    # resort to stick back in matrix
    resorted_list = []
    for i in author_list:
        if i == curr_name:
            resorted_list.append(20)
        else:
            resorted_list.append(curr_similarities[curr_names_sorted.index(i)]+10)
    # putting in matrix
    all_cos_sim[author_list.index(curr_name),:] = resorted_list
```

```python
plt.imshow(all_cos_sim)
plt.colorbar()
plt.show()
# np.log10(10)
```

```python
# now lets make a graph
G = nx.DiGraph()
G.add_nodes_from(author_list)
for i in range(0,len(author_list)):
    for j in range(i,len(author_list)): 
        if all_cos_sim[i,j] > 2:
            G.add_edge(author_list[i], author_list[j], weight=np.max([all_cos_sim[i,j], 0]))
        else:
            G.add_edge(author_list[i], author_list[j], weight=1)
```

```python
# for visualization
pos = nx.kamada_kawai_layout(G)
# pos=nx.spring_layout(G, iterations=3000)
# pos=nx.circular_layout(G)
cm = plt.get_cmap('inferno_r', 7) # 7 bc of the 5 positions and no black
unique_pos = ['PG', 'SG', 'SF', 'PF', 'C']
curr_positions = nx.get_node_attributes(G, 'pos')
nodecolors = [cm.colors[j] for i in curr_positions for j in range(0,len(unique_pos))
              if curr_positions[i] == unique_pos[j]]
edge_labels=dict([((u, v, ), round(d['weight']))
                 for u, v, d in G.edges(data=True)])
# for the node size
node_size_use = nx.get_node_attributes(G, 'ppg')
node_array = [node_size_use[i]*25 for i in list(node_size_use)]
labels_use = {}
for i in G.nodes():
    labels_use[i] = i
# print(edge_labels)
edgewidth = [d['weight']/40 for (u, v, d) in G.edges(data=True)]
plt.figure(figsize=(10, 4))
# # we can now added edge thickness and edge color
nx.draw_networkx_nodes(G, pos=pos, alpha=0.8)
nx.draw_networkx_edges(G, pos, width=edgewidth, alpha=0.15, arrowstyle='simple',
                       arrowsize=5, arrows=False, edge_cmap=plt.cm.Greys)
nx.draw_networkx_labels(G, pos, labels_use, font_size=8)
plt.axis('off')
plt.tight_layout()
```
