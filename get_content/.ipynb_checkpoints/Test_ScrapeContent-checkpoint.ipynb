# Scrape Articles
Using the url from muckrack.com, go to each website and scrape all of the articles and push to MongoDB. Within my database I will have a collection for each writer that has all information for a given article including the text of the article. I am using MongoDB because it does well with unstructured information

```python
# load in information about url that I saved from muckrack.com
with open('/Users/rohanramesh/Documents/Insight/data_bball_writers/Attempt4_mr_scrape.pickle', 'rb') as handle:
    scrapevar = pickle.load(handle)
```

```python
# Lets visualize how many articles I have from each site
all_sources = []
for curr_name in scrapevar['sources']:
    for curr_source in scrapevar['sources'][curr_name]:
        all_sources.append(curr_source)

source_counts = Counter(all_sources)
df = pd.DataFrame.from_dict(source_counts, orient='index')
col = list(df)

# Plot top 10 number of articles
a = df.sort_values(by=col[0],ascending=False)
name_sites_sorted = list(a[0].index)
narticles_sites_sorted = list(a[0].values)
nplot = 20
plt.bar(name_sites_sorted[0:nplot], narticles_sites_sorted[0:nplot], align='center')
plt.xticks(rotation=60)
plt.ylabel('Number of articles')
```

```python
# grab relevant information for article
def grab_article_info(curr_name,link_idx, scrapevar):
    """
    from dict with muckrack information, get the relevan url and article source
    :param curr_name: name of writer in muckrack from is first_name-last_name
    :param link_idx: which number back in time the article was ie the fifth most recent article
    :param scrapevar: dict from scraping muckrack
    :return: article_url, article_source, article_pubdate: source is website, others self explanatory
    """
    article_url = scrapevar['links'][curr_name][link_idx]
    article_source = scrapevar['sources'][curr_name][link_idx]
    article_pubdate = scrapevar['pubdates'][curr_name][link_idx]
    return article_url, article_source, article_pubdate
```

```python
# initialize with mongo
client = MongoClient()
# try building mongodb
mydb = client["testinsightdb"]
# mydb = client["insight_wtitle"]
```

```python
# load writer df
with open('/Users/rohanramesh/Documents/Insight/data_bball_writers/writer_df.pickle', 'rb') as handle:
    writer_df = pickle.load(handle)
```

```python
# iterate through all writer names and scrape all articles from websites
for curr_name in writer_df['website_name']:
    # later useful for saving full name from writer_df
    idx = writer_df['website_name'] == curr_name
    print(curr_name+' '+str(datetime.datetime.now()))
    # initialize new collection
    mycol = mydb[curr_name]
    rr = []
    idfail = 1
    # iterate through all links from muckrack
    for link_idx in range(0,len(scrapevar['links'][curr_name])):
        # get relevant article info from muckrack information
        article_url, article_source, article_pubdate = grab_article_info(curr_name,link_idx, scrapevar)
        # try downloading article and if fail move on
        try:
            # using newspaper3k to download articles
            article = Article(article_url)
            article.download()
            article.parse()
            article.text
            # randomize wait time so they don't kick me off
            time.sleep(random.uniform(0.5,1))
            # build dict with relevant information and push to db
            mydict = { "name": curr_name, "article_url": article_url, "article_source": article_source, 
                     "pubdates": article_pubdate,"writer_id": writer_df['Idea ID'][idx].tolist()[0], 
                      "article": article.text, "authors": article.authors, "narticle": link_idx, 
                     "article_summary": article.summary, "article_title": article.title}
            x = mycol.insert_one(mydict)
        except Exception:
            # after fail 30 times move on to next writer
            print(curr_name + ' ' + str(link_idx))
            idfail = idfail + 1
            print('fail # ' + str(idfail))
            if idfail > 29:
                print('breaking')
                break
            else:
                continue
```
